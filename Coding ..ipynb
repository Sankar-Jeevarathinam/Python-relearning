```python
#5.Write a Python program to check if a number is prime. Expected Output (for input 7):
#7 is a prime number
import math
def is_prime():
    num=int(input("Enter number="))
    if num<=1:
        msg=False
    elif (num==2):
        msg=True
    else:
        msg=True
        for i in range (2, int(math.sqrt(num)+1)):
            if ((num % i)==0):
                msg=False
                break

    return msg
```

```python
msg=is_prime()
if msg:
    print ("The number is Prime")
else:
    print ("The number is not Prime")
```
